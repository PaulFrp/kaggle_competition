```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("dsba-m-1-challenge-purchase-prediction/train_dataset_M1_with_id.csv")
```

```python
print(df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
print(df.isnull().sum().sort_values(ascending=False))
```

```python
#This tells us the proportion of missing values for each feature
print(df.isnull().sum().sort_values(ascending=False) / len(df))
```

```python
#We have 0 duplicated values to start with in the train dataset
print(df.duplicated().sum())
```

```python
# Now I check for outliers and plot the distributions of each feature
numerical_data = df.select_dtypes(include=['int64', 'float64'])
for feature in numerical_data:
    plt.figure()
    sns.histplot(data=df, x=feature, kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
 # Checking for correlation between features
 #It seems only purchase and items in cart are correlated so it is hard to do some feature engineering just now
corr_matrix = numerical_data.corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
df[['Payment_Method', 'Referral_Source2']] = df['PM_RS_Combo'].str.split(':', expand=True)
```

```python
# Average discount per referral source
avg_discount_per_source = (
    df.groupby("Referral_Source")["Discount"]
    .mean()
    .sort_values(ascending=False)
)
print("Average Discount per Referral Source:")
print(avg_discount_per_source)
```

```python
# Conversion rate (purchase rate) by referral source
conversion_rate_by_referral = (
    df.groupby("Referral_Source")["Purchase"]
    .mean()
    .sort_values(ascending=False)
)
print("Conversion Rate by Referral Source:")
print(conversion_rate_by_referral)
```

```python
# Average items in cart by category
avg_items_in_cart_by_category = (
    df.groupby("Category")["Items_In_Cart"]
    .mean()
    .sort_values(ascending=False)
)
print("Average Items in Cart by Category:")
print(avg_items_in_cart_by_category)
```
